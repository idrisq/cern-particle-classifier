```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Notebook içinde grafiklerin düzgün görüntülenmesi için
%matplotlib inline

# Grafiklerin daha okunaklı olması için bir stil belirleyelim
sns.set_style('whitegrid')
```

```python
# Veri setinin yolunu tanımlıyoruz. ../ bir üst klasöre çık demek.
DATA_PATH = '../data/raw/training.csv'

# CSV dosyasını okuyup bir DataFrame oluşturuyoruz.
df = pd.read_csv(DATA_PATH)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='Label', data=df)
plt.title('Sinyal ve Arka Plan Olay Sayıları')
plt.show()
print(df['Label'].value_counts())
```

```python
# -999.0 değerlerinin grafikleri bozmaması için onları geçici olarak NaN ile değiştiriyoruz.
# Orijinal df'i korumak adına yeni bir dataframe oluşturmak iyi bir pratiktir.
df_vis = df.replace(-999.0, np.nan)
```

```python
# 'Label' sütunundaki 's' ve 'b' değerlerini daha anlamlı etiketlerle değiştirelim.
# Bunu yeni bir sütuna atayarak orijinal veriyi koruyoruz.
label_map = {'s': 'Sinyal', 'b': 'Arka Plan'}
df_vis['Olay Tipi'] = df_vis['Label'].map(label_map)

# Yeni sütunun doğru oluştuğunu kontrol edelim
df_vis[['Label', 'Olay Tipi']].head()
```

```python
# GÜNCELLENMİŞ VE DOĞRU GRAFİK KODU

plt.figure(figsize=(12, 8))

# 'hue' parametresinde artık yeni oluşturduğumuz 'Olay Tipi' sütununu kullanıyoruz.
sns.histplot(data=df_vis, x='DER_mass_MMC', hue='Olay Tipi', 
             stat='density', common_norm=False, bins=50, kde=True)

plt.title('DER_mass_MMC Dağılımı (Sinyal vs Arka Plan)', fontsize=16)
plt.xlabel('Tahmini Kütle (DER_mass_MMC)', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)

# Sorunlu olan plt.legend() satırını sildik. Seaborn artık doğru etiketi otomatik olarak oluşturacak.
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=df_vis, x='PRI_tau_pt', hue='Olay Tipi', 
             stat='density', common_norm=False, bins=50, kde=True)
plt.title('PRI_tau_pt Dağılımı (Sinyal vs Arka Plan)', fontsize=16)
plt.xlabel('Tau Leptonu Enine Momentum (PRI_tau_pt)', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=df_vis, x='DER_mass_jet_jet', hue='Olay Tipi', 
             stat='density', common_norm=False, bins=50, kde=True)
plt.title('DER_mass_jet_jet Dağılımı (Sinyal vs Arka Plan)', fontsize=16)
plt.xlabel('İki Jet Arası Kütle (DER_mass_jet_jet)', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
plt.show()
```
